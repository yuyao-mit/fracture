```python
import sys
import torch
sys.path.append("./neuraloperator")
from neuralop.models.codano import CODANO
```

```python


# 假设是 2D PDE 输入（例如流体力学）
batch_size = 2
num_vars = 3  # 输入物理变量个数，例如 [u_x, u_y, p]
height, width = 64, 64

# 构造随机输入
x = torch.randn(batch_size, num_vars, height, width)

# 初始化 CODANO 模型
model = CODANO(
    output_variable_codimension=1,       # 每个变量对应 1 个输出通道
    lifting_channels=64,
    hidden_variable_codimension=32,
    projection_channels=64,
    n_layers=4,
    n_modes=[[16, 16]] * 4,              # 每层 Fourier 模式数
    n_heads=[2, 2, 2, 2],                # 每层注意力头数
    per_layer_scaling_factors=[[1, 1]] * 4,
    attention_scaling_factors=[1] * 4,
    use_positional_encoding=True,        # 如果需要位置编码
    positional_encoding_dim=8,
    positional_encoding_modes=[16, 16],  # Fourier 模式数（与 n_modes 对应）
    variable_ids=["u_x", "u_y", "p"],    # 变量名要和输入通道对应
)

# 前向调用
y = model(x, input_variable_ids=["u_x", "u_y", "p"])

print(f"输入形状: {x.shape}")
print(f"输出形状: {y.shape}")
```
